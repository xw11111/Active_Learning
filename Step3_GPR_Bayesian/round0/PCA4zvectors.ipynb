```python
import numpy as np
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

# Load the data from the .npy file
data = np.load('../Step2_RAE_embedding/hyperparameters_tuned/A_Nnum_Sep/z_vectors.npy')  # Replace 'data.npy' with your file path
print(np.shape(data))

# Perform k-means clustering
k = 100  # Number of clusters, you can adjust this number as needed
kmeans = KMeans(n_clusters=k, random_state=0, n_init=10).fit(data)

# Get cluster centroids
centroids = kmeans.cluster_centers_

# Find the points closest to each cluster's centroid
distances = cdist(data, centroids, 'euclidean')
indices_round0 = np.array([18, 61, 115, 133, 242, 671, 889, 957, 1027, 1034, 1042, 1065,
                  1103, 1364, 1511, 1518, 1658, 2068, 2093, 2123, 2423, 2466, 
                  2477, 2678, 2750, 3078, 4121, 4272, 4800, 4808, 5323, 5421, 
                  5517, 5584, 5587, 5637, 5654, 5704, 5712, 5715, 5733, 5756, 
                  5884, 6133, 6222, 6334, 6383, 6415, 6479, 6608, 6648, 6884, 
                  7198, 7254, 7792, 8172, 8212, 8240, 8250, 8267, 8435, 8493, 
                  8631, 8664, 8691, 8987, 9023, 9267, 9538, 9556, 9696, 9708, 
                  9711, 9727, 9729, 9769, 9806, 9819, 9879, 10483, 10653, 
                  10710, 10736, 10742, 10760, 10887, 10889, 10897, 10960, 
                  11020, 11072, 11540, 11610, 11636, 11668, 11804, 11900, 
                  12018, 12055, 12094])

# Get the points closest to each centroid
round0_points = data[indices_round0]

# Display the results
# print("Cluster centroids:\n", centroids)
print("Indices of points closest to each centroid:\n", indices_round0)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the 16-dimensional vectors
allzvector = data

# Perform PCA to reduce the dimension to 2D
pca = PCA(n_components=2)
reduced_allzvector = pca.fit_transform(allzvector)
# save the PCAed data
np.save("PCAed_Zvectors.npy", np.array(reduced_allzvector))

# Plot the reduced 2D vectors
plt.figure(figsize=(10, 6))
plt.scatter(reduced_allzvector[:, 0], reduced_allzvector[:, 1], color='cyan', label='all data')
plt.title('PCA Reduction to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Example 2D arrays
zvector4bds = reduced_allzvector[:10902]
zvector3bds = reduced_allzvector[10902:12028]
zvector2bds = reduced_allzvector[12028:12122]
zvector1bd = reduced_allzvector[12122:]

# Create a scatter plot
plt.figure(figsize=(10, 6))

# Plot 4 beads
plt.scatter(zvector4bds[:, 0], zvector4bds[:, 1], color='blue', label='4 beads')

# Plot 3 beads
plt.scatter(zvector3bds[:, 0], zvector3bds[:, 1], color='red', label='3 beads')

# Plot 2 beads
plt.scatter(zvector2bds[:, 0], zvector2bds[:, 1], color='green', label='2 beads')

# Plot 1 beads
plt.scatter(zvector1bd[:, 0], zvector1bd[:, 1], color='cyan', label='1 beads')

# Adding titles and labels
plt.title('Scatter Plot of 4 lengths of beads')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
